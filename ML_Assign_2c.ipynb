```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve,auc
from sklearn.model_selection import KFold
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve, classification_report, precision_recall_curve
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, GridSearchCV, RandomizedSearchCV
```

```python
dataset = pd.read_csv("/content/drug_dataset.csv")
```

```python
dataset.head()
```

```python
dataset.drop(columns=['ID'])
```

```python
df = dataset.copy()
```

```python
columns = ['VSA']
```

```python
from sklearn.preprocessing import LabelEncoder
for column in columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
```

```python
for i in range(len(columns)):
    df.loc[((df[columns[i]]==0) | (df[columns[i]]==1)),columns[i]] = 'Non-user'
    df.loc[((df[columns[i]]==2) | (df[columns[i]]==3) | (df[columns[i]]==4) | (df[columns[i]]==5) | (df[columns[i]]==6)),columns[i]] = 'User'
```

```python
count_of_users = []
count_of_non_users = []
```

```python
for i in range(len(columns)):
    s = df.groupby([columns[i]])[columns[i]].count()
    count_of_users.append(s[1])
    count_of_non_users.append(s[0])
```

```python
x=count_of_users
y=count_of_non_users
print(x)
print(y)
```

```python
for column in columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
```

```python
features = ['Age', 'Gender', 'Education', 'Country', 'Ethnicity', 'Nscore',
       'Escore', 'Oscore', 'Ascore', 'Cscore', 'Impulsive', 'SS']
target = ['VSA']

X = df[features]
y = df[target]
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression
X = SequentialFeatureSelector(estimator=LogisticRegression(), n_features_to_select = 9, cv = 10, direction ='backward').fit_transform(X,y)
```

```python
X[0] #Gender, E-Score and C-Score was removed in feature selection
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
X[0]
```

#UNDER_SAMPLING

#KNN 

```python
#K-Fold Cross Validation
from sklearn.neighbors import KNeighborsClassifier
rus = RandomUnderSampler(random_state=0)
acc_score = []
acc_train = []

kf = KFold(n_splits=10, random_state=42, shuffle=True)
knn = KNeighborsClassifier(metric = "minkowski", n_neighbors = 21, algorithm='ball_tree',weights='uniform')

for train_index, test_index in kf.split(X):
  X_train, X_test = X[train_index,:], X[test_index,:]
  y_train, y_test = y.loc[train_index], y.loc[test_index]
  X_us, y_us = rus.fit_resample(X_train, y_train)

  knn.fit(X_us,y_us)
  training1 = knn.predict(X_us)
  pred1 = knn.predict(X_test)

  acc = accuracy_score(pred1, y_test)
  acc_score.append(acc)
  acc1 = accuracy_score(training1,y_us)
  acc_train.append(acc1)


avg_acc = sum(acc_score)/10
avg_train_acc= sum(acc_train)/10

print('Accuracy in 10 folds Training - {}'.format(acc_train))
print('AVG Accuracy Training: {}'.format(avg_train_acc))

print('Accuracy in 10 folds - {}'.format(acc_score))
print('AVG Accuracy : {}'.format(avg_acc))
```

#SVM 

```python
#K-Fold Cross Validation
from sklearn.metrics.pairwise import rbf_kernel
from sklearn.svm import SVC
rus = RandomUnderSampler(random_state=0)
acc_score = []
acc_train = []

kf = KFold(n_splits=10, random_state=42, shuffle=True)
svm=SVC(C=0.1, gamma=1, kernel='poly')

for train_index, test_index in kf.split(X):
  X_train, X_test = X[train_index,:], X[test_index,:]
  y_train, y_test = y.loc[train_index], y.loc[test_index]
  X_us, y_us = rus.fit_resample(X_train, y_train)

  svm.fit(X_us,y_us)
  training2 = svm.predict(X_us)
  pred2 = svm.predict(X_test)

  acc = accuracy_score(pred2, y_test)
  acc_score.append(acc)
  acc1 = accuracy_score(training2, y_us)
  acc_train.append(acc1)


avg_acc = sum(acc_score)/10
avg_train_acc= sum(acc_train)/10

print('Accuracy in 10 folds Training - {}'.format(acc_train))
print('AVG Accuracy Training: {}'.format(avg_train_acc))

print('Accuracy in 10 folds - {}'.format(acc_score))
print('AVG Accuracy : {}'.format(avg_acc))
```

#DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
rus = RandomUnderSampler(random_state=0)
acc_score = []
acc_train = []

kf = KFold(n_splits=10, random_state=42, shuffle=True)
dt = DecisionTreeClassifier(max_depth=4, min_samples_leaf=5, random_state=42, criterion='entropy')

for train_index, test_index in kf.split(X):
  X_train, X_test = X[train_index,:], X[test_index,:]
  y_train, y_test = y.loc[train_index], y.loc[test_index]
  X_us, y_us = rus.fit_resample(X_train, y_train)

  dt.fit(X_us,y_us)
  training3 = dt.predict(X_us)
  pred3 = dt.predict(X_test)

  acc = accuracy_score(pred3, y_test)
  acc_score.append(acc)
  acc1 = accuracy_score(training3, y_us)
  acc_train.append(acc1)


avg_acc = sum(acc_score)/10
avg_train_acc= sum(acc_train)/10

print('Accuracy in 10 folds Training - {}'.format(acc_train))
print('AVG Accuracy Training: {}'.format(avg_train_acc))

print('Accuracy in 10 folds - {}'.format(acc_score))
print('AVG Accuracy : {}'.format(avg_acc))
```
