```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve,auc
from sklearn.model_selection import KFold
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve, classification_report, precision_recall_curve
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, GridSearchCV, RandomizedSearchCV
```

```python
dataset = pd.read_csv("/content/drug_dataset.csv")
```

```python
dataset.head()
```

```python
dataset.drop(columns=['ID'])
```

```python
df = dataset.copy()
```

```python
columns = ['VSA']
```

```python
from sklearn.preprocessing import LabelEncoder
for column in columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
```

```python
for i in range(len(columns)):
    df.loc[((df[columns[i]]==0) | (df[columns[i]]==1)),columns[i]] = 'Non-user'
    df.loc[((df[columns[i]]==2) | (df[columns[i]]==3) | (df[columns[i]]==4) | (df[columns[i]]==5) | (df[columns[i]]==6)),columns[i]] = 'User'
```

```python
count_of_users = []
count_of_non_users = []
```

```python
for i in range(len(columns)):
    s = df.groupby([columns[i]])[columns[i]].count()
    count_of_users.append(s[1])
    count_of_non_users.append(s[0])
```

```python
x=count_of_users
y=count_of_non_users
print(x)
print(y)
```

```python
for column in columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
```

```python
features = ['Age', 'Gender', 'Education', 'Country', 'Ethnicity', 'Nscore',
       'Escore', 'Oscore', 'Ascore', 'Cscore', 'Impulsive', 'SS']
target = ['VSA']

X = df[features]
y = df[target]
```

Feature selection using Logistic Regression

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression
X = SequentialFeatureSelector(estimator=LogisticRegression(), n_features_to_select = 9, cv = 10, direction ='backward').fit_transform(X,y)
```

```python
X[0] #Gender, E-Score and C-Score was removed in feature selection
```

Feature Scaling Using MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
X[0]
```

#OVERSAMPLING

For Oversampling used the SMOTE Technique on the training folds of crossvalidation 

#KNN 

```python
#K-Fold Cross Validation
from sklearn.neighbors import KNeighborsClassifier
ovs = SMOTE(random_state=42)
acc_score = []
acc_train = []

kf = KFold(n_splits=10, random_state=42, shuffle=True)
knn = KNeighborsClassifier(metric = "manhattan", n_neighbors = 2, algorithm='ball_tree',weights='uniform')

for train_index, test_index in kf.split(X):
  X_train, X_test = X[train_index,:], X[test_index,:]
  y_train, y_test = y.loc[train_index], y.loc[test_index]
  X_ovs, y_ovs = ovs.fit_resample(X_train,y_train)

  knn.fit(X_ovs,y_ovs)
  y_training1 = knn.predict(X_ovs)
  y_pred1 = knn.predict(X_test)

  acc = accuracy_score(y_test,y_pred1)
  acc_score.append(acc)
  acc1 = accuracy_score(y_ovs,y_training1)
  acc_train.append(acc1)


avg_acc = sum(acc_score)/10
avg_train_acc= sum(acc_train)/10

print('Accuracy in 10 folds Training - {}'.format(acc_train))
print('AVG Accuracy Training: {}'.format(avg_train_acc))

print('Accuracy in 10 folds - {}'.format(acc_score))
print('AVG Accuracy : {}'.format(avg_acc))
```

#SVM 

```python
#K-Fold Cross Validation
from sklearn.metrics.pairwise import rbf_kernel
from sklearn.svm import SVC
ovs = SMOTE()
acc_score = []
acc_train = []

kf = KFold(n_splits=10, random_state=42, shuffle=True)
svm=SVC(C=1000, gamma=1, kernel='rbf')

for train_index, test_index in kf.split(X):
  X_train, X_test = X[train_index,:], X[test_index,:]
  y_train, y_test = y.loc[train_index], y.loc[test_index]
  X_ovs, y_ovs = ovs.fit_resample(X_train,y_train)

  svm.fit(X_ovs,y_ovs)
  y_training2 = svm.predict(X_ovs)
  y_pred2 = svm.predict(X_test)

  acc = accuracy_score(y_test,y_pred2)
  acc_score.append(acc)
  acc1 = accuracy_score(y_ovs,y_training2)
  acc_train.append(acc1)


avg_acc = sum(acc_score)/10
avg_train_acc= sum(acc_train)/10

print('Accuracy in 10 folds Training - {}'.format(acc_train))
print('AVG Accuracy Training: {}'.format(avg_train_acc))

print('Accuracy in 10 folds - {}'.format(acc_score))
print('AVG Accuracy : {}'.format(avg_acc))
```

#DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
ovs = SMOTE()
acc_score = []
acc_train = []

kf = KFold(n_splits=10,random_state=42, shuffle=True)
dt = DecisionTreeClassifier(max_depth=12, min_samples_leaf=5, random_state=42, criterion='entropy')

for train_index, test_index in kf.split(X):
  X_train, X_test = X[train_index,:], X[test_index,:]
  y_train, y_test = y.loc[train_index], y.loc[test_index]
  X_ovs, y_ovs = ovs.fit_resample(X_train,y_train)

  dt.fit(X_ovs,y_ovs)
  y_training3 = dt.predict(X_ovs)
  y_pred3 = dt.predict(X_test)

  acc = accuracy_score(y_test,y_pred3)
  acc_score.append(acc)
  acc1 = accuracy_score(y_ovs,y_training3)
  acc_train.append(acc1)


avg_acc = sum(acc_score)/10
avg_train_acc= sum(acc_train)/10

print('Accuracy in 10 folds Training - {}'.format(acc_train))
print('AVG Accuracy Training: {}'.format(avg_train_acc))

print('Accuracy in 10 folds - {}'.format(acc_score))
print('AVG Accuracy : {}'.format(avg_acc))
```

#RANDOM FOREST

```python
#K-Fold Cross Validation
from sklearn.ensemble import RandomForestClassifier
ovs = SMOTE()
acc_score = []
acc_train = []

kf = KFold(n_splits=10, random_state=42, shuffle = True)
rf = RandomForestClassifier(n_estimators=100,random_state=0)

for train_index, test_index in kf.split(X):
  X_train, X_test = X[train_index,:], X[test_index,:]
  y_train, y_test = y.loc[train_index], y.loc[test_index]
  X_ovs, y_ovs = ovs.fit_resample(X_train,y_train)

  rf.fit(X_ovs,y_ovs)
  y_training4 = rf.predict(X_ovs)
  y_pred4 = rf.predict(X_test)

  acc = accuracy_score(y_test,y_pred4)
  acc_score.append(acc)
  acc1 = accuracy_score(y_ovs,y_training4)
  acc_train.append(acc1)


avg_acc = sum(acc_score)/10
avg_train_acc= sum(acc_train)/10

print('Accuracy in 10 folds Training - {}'.format(acc_train))
print('AVG Accuracy Training: {}'.format(avg_train_acc))

print('Accuracy in 10 folds - {}'.format(acc_score))
print('AVG Accuracy : {}'.format(avg_acc))
```

#MULTILAYER PERCEPTRON

```python
#K-Fold Cross Validation
from sklearn.neural_network import MLPClassifier
ovs = SMOTE()
acc_score = []
acc_train = []

kf = KFold(n_splits=10, random_state=42, shuffle = True)
clf = MLPClassifier(alpha=0.0001,learning_rate='constant',solver='adam',activation='relu',hidden_layer_sizes=(150, 100, 50))

for train_index, test_index in kf.split(X):
  X_train, X_test = X[train_index,:], X[test_index,:]
  y_train, y_test = y.loc[train_index], y.loc[test_index]
  X_ovs, y_ovs = ovs.fit_resample(X_train,y_train)

  clf.fit(X_ovs,y_ovs)
  y_training5 = clf.predict(X_ovs)
  y_pred5 = clf.predict(X_test)

  acc = accuracy_score(y_test,y_pred5)
  acc_score.append(acc)
  acc1 = accuracy_score(y_ovs,y_training5)
  acc_train.append(acc1)


avg_acc = sum(acc_score)/10
avg_train_acc= sum(acc_train)/10

print('Accuracy in 10 folds Training - {}'.format(acc_train))
print('AVG Accuracy Training: {}'.format(avg_train_acc))

print('Accuracy in 10 folds - {}'.format(acc_score))
print('AVG Accuracy : {}'.format(avg_acc))
```
